```python
#importing necessary packages
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Generating uniform random number and plotting
np.random.seed(19680801) #starting seed to generate random numbers
N = 50 #number of random numbers generated
#x and y values between 0 and 1, they are arrays of size N
x = np.random.rand(N)
y = np.random.rand(N)
#defining color array
colors = np.random.rand(N)
#defining area of the points in terms of the array
area = (30 * np.random.rand(N))**2  
#plotting 
plt.scatter(x, y, s=area, c=colors)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()

```

```python
#drawing gaussian random numbers
mu=2
sigma=0.2
N=10**5
#N gaussian random numbers 
x1=np.random.normal(mu, sigma, N)
xmin=mu-5*sigma
xmax=mu+5*sigma
dx=0.75*sigma
x=np.linspace(xmin, xmax, 100) #100 equally spaced x points
y=N*np.exp(-0.5*(x-mu)**2/sigma**2)/(sigma*(2*math.pi)**0.5)*dx
#plotting x1
plt.hist(x1,width=dx,facecolor='g',edgecolor='k')
plt.plot(x,y,'r-',linewidth=2)
plt.title("Gaussian", fontsize=20, fontweight='bold')
plt.show()
```

```python
#Generating poisson random variables
mu=5
x2=np.random.poisson(mu, 100000) #mu is mean
#plotting x1
plt.hist(x2,15,facecolor='r',edgecolor='k')
plt.title("Poisson", fontsize=20, fontweight='bold')
plt.show()
```

```python
#Generating binomial random variables
mu=5
x2=np.random.binomial(1000,0.2,10^5)
#plotting x1
plt.hist(x2,15,facecolor='m',edgecolor='k')
plt.title("Binomial", fontsize=20, fontweight='bold')
plt.show()
```
