```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
#x and y are x and y coordinates, z is size, 'green is color'
plt.scatter(x, y, z, 'green')
plt.show()
```

```python
#colors can also be represented by numbers
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.show()
```

```python
#x,ylabel,title
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs. x")
plt.show()
```

```python
#x,ylabel,title, font size
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs. x")
plt.rc('font', size=10)
plt.show()
```

```python
#x,ylabel,title, font size
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs. x")
plt.rc('font', size=5)
plt.show()
```

```python
#saving figure
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs. x")
plt.rc('font', size=15)
plt.savefig("test.png")
plt.show()
```

```python
#Different fontsize for different elements
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.show()
```

```python
#grid on
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()
```

```python
#x and y ticks 
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.xticks([1,3,5,7], fontsize=10)
plt.yticks([2,4,6], fontsize=20)
plt.show()
```

```python
#x and y ticklabels
x=[1,2,4,5,2,5,2,5,7]
y=[2,3,5,3,4,4,5,6,7]
z=[n**2 for n in x]
plt.scatter(x, y, z, z)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.xticks([1,3,5,7], fontsize=10)
plt.yticks([2,4,6],["low","medium","high"],fontsize=20)
plt.show()
```

```python
#new plotstyles
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'go-')
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()
```

```python
#new plotstyles: markerfacecolor and markersize
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'o-', markersize=8, markerfacecolor='r')
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()
```

```python
#new plotstyles markeredgecolor
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'o-', markersize=12, markerfacecolor='r',markeredgewidth=2,  markeredgecolor='k')
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()

```

```python
#new plotstyles linewidth and linecolor
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor='r',markeredgewidth=2,  markeredgecolor='k',linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()
```

```python
#creating your own color
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.grid()
plt.show()
```

```python
#Setting axes to log scale 
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.yscale('log')
plt.grid()
plt.show()
```

```python
#Setting axes  limits 
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20)
plt.yscale('log')
plt.grid()
plt.xlim([2,4]) #x limit
plt.show()
```

```python
#Setting title font to bold
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20, fontweight='bold')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
#Setting axes line width
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20, fontweight='bold')
ax = plt.gca()
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(3)
plt.grid()
plt.show()
```

```python
#Setting legends
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
z2=[n**3 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.plot(z1, z2,'cs-', markersize=12, markerfacecolor='g',markeredgewidth=2,  markeredgecolor='k', linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20, fontweight='bold')
ax = plt.gca()
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(3)
plt.yscale('log')
plt.legend(['y=x^2','y=x^3'])
plt.grid()
plt.show()
```

```python
#Setting legends location
c1=[0.65, 0, 0.25] #new color 1
c2=[1, 0.5, 0.5] #new color 2
z1=[1,2,3,4,5]
z=[n**2 for n in z1]
z2=[n**3 for n in z1]
plt.plot(z1, z,'mo-', markersize=12, markerfacecolor=c1,markeredgewidth=2,  markeredgecolor=c2, linewidth=2)
plt.plot(z1, z2,'cs-', markersize=12, markerfacecolor='g',markeredgewidth=2,  markeredgecolor='k', linewidth=2)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=15)
plt.title("y vs. x", fontsize=20, fontweight='bold')
ax = plt.gca()
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(3)
plt.yscale('log')
plt.legend(['y=x^2','y=x^3'], loc='lower right')
plt.grid()
plt.show()
```
